```python
import pandas as pd
import numpy as np
import json
from google.colab import drive
drive.mount('/content/drive')
dir = '/content/drive/MyDrive/analytics/fewshot_medical'
```

```python
train_test_ratios = [0.8, 0.7, 0.6, 0.5, 0.4, 0.3]
embedding_modes = ['fasttext', 'bert']
```

```python
way = 19

result_compilation = []

for embedding_mode in embedding_modes:
    results = {}
    for train_test_ratio in train_test_ratios:
        result = pd.read_csv(f'{dir}/train_test_ratio/{way}_way/{train_test_ratio}/{embedding_mode}/performance_result.csv')['total_f1_score_percent'][0]
        results[train_test_ratio] = result
    result_compilation.append(results)

result_df_19_way = pd.DataFrame(result_compilation)
result_df_19_way.index = embedding_modes
```
